Now that the text data is ready we process the downstream tasks to work with the text classification scripts. 

```python
from asyncore import write
import os
import argparse
import json
import numpy
from sklearn.preprocessing import StandardScaler
def process_ner_dataset(filepath, target_filepath):
    sentences = []
    with open(filepath,'r') as f:
        current = []
        for l in f:
            if l == '\n' and current != []:
                if '' in current:
                    current.remove('')
                sentences.append(current)
                current = []
            elif '-DOCSTART-' in l:
                pass
            else:
                current.append(l.strip())
    dataset = []
    for sentence in sentences:
        item = {"words":[], "ner":[]}
        for segment in sentence:
            word, tag = segment.split('\t')
            item["words"].append(word)
            item["ner"].append(tag)
        dataset.append(item)
    with open(target_filepath,'w') as w:
        for item in dataset:
            w.write("{}\n".format(json.dumps(item)))

def process_hoc(filepath, target_filepath):
    label_lookup = ["PS","GS","CD","RI","A","IM","GI","TPI","CE","ID"]
    with open(filepath,'r') as f:
        f.readline()
        with open(target_filepath,'w') as w:
            for l in f:
                label, sentence, guid = l.strip().split('\t')
                labels = []
                text_labels = []
                for item in label.split(','):
                   labels.append(float(item[2]))
                   if int(item[2]) > 0:
                       text_labels.append(label_lookup[int(item[0])])
                w.write("{}\n".format(json.dumps({'sentence':sentence, 'text_labels': text_labels,'label': labels, 'guid': guid})))

def process_bioses(filepath):
    x, y = [], []
    with open(filepath,'r') as f:
        f.readline()
        for l in f:
            guid, sentence1, sentence2,  label = l.strip().split('\t')
            x.append({'id': guid, 'sentence1':sentence1, 'sentence2':sentence2,"original_label": float(label), "label": float(label)})
            y.append(float(label))
    return x,numpy.asarray(y)
        
def process_pubmedqa(filepath, target_filepath):
    with open(filepath,'r') as f:
        dataset = json.load(f)
        with open(target_filepath,'w') as w:
            for item_id in dataset:
                query = dataset[item_id]['QUESTION']
                contexts = '. '.join(dataset[item_id]["CONTEXTS"])
                label = dataset[item_id]["final_decision"]
                w.write("{}\n".format(json.dumps({'id': item_id, 'query':query, 'context':contexts, 'label': label})))

def process_bioasq(filepath, target_filepath):
    with open(filepath,'r') as f:
        with open(target_filepath, 'w') as w:
            for l in f:
                l = l.strip().split('\t')
                query = l[1]
                item_id = l[0]
                context = l[2]
                label = l[3]
                w.write("{}\n".format(json.dumps({'id': item_id, 'query':query, 'context':context, 'label': label})))

def process_GAD(filepath, target_filepath):
    with open(filepath,'r') as f:
        with open(target_filepath, 'w') as w:
            for l in f:
                l = l.strip().split('\t')
                item_id = l[0]
                sentence = l[1]
                label = l[2]
                w.write("{}\n".format(json.dumps({'id': item_id, 'sentence': sentence, 'label': label})))

def process_chemprot(filepath, target_filepath):
    with open(filepath,'r') as f:
        with open(target_filepath, 'w') as w:
            for l in f:
                l = l.strip().split('\t')
                sentence = l[1]
                label = l[2]
                w.write("{}\n".format(json.dumps({ 'sentence': sentence, 'label': label})))

def process_ddi(filepath, target_filepath):
    with open(filepath,'r') as f:
        f.readline()
        with open(target_filepath, 'w') as w:
            for l in f:
                l = l.strip().split('\t')
                sentence = l[1]
                label = l[2]
                w.write("{}\n".format(json.dumps({ 'sentence': sentence, 'label': label})))

def print_biosses(x,y,filepath):
    with open(filepath, 'w') as w:
        for i in range(len(y)):
            x[i]['remapped_label'] = float(y[i])
            w.write("{}\n".format(json.dumps(x[i])))


print("Processing NER Datasets")
print("Processing the BC2GM Dataset")
process_ner_dataset('data/BC2GM/train.tsv', 'data/BC2GM/train.json')
process_ner_dataset('data/BC2GM/dev.tsv', 'data/BC2GM/dev.json')
process_ner_dataset('data/BC2GM/test.tsv', 'data/BC2GM/test.json')

print("Processing BC5CDR-chem")
process_ner_dataset('data/BC5CDR-chem/train.tsv', 'data/BC5CDR-chem/train.json')
process_ner_dataset('data/BC5CDR-chem/dev.tsv', 'data/BC5CDR-chem/dev.json')
process_ner_dataset('data/BC5CDR-chem/test.tsv', 'data/BC5CDR-chem/test.json')
print("Processing BC5CDR-disease")
process_ner_dataset('data/BC5CDR-disease/train.tsv', 'data/BC5CDR-disease/train.json')
process_ner_dataset('data/BC5CDR-disease/dev.tsv', 'data/BC5CDR-disease/dev.json')
process_ner_dataset('data/BC5CDR-disease/test.tsv', 'data/BC5CDR-disease/test.json')

print("Processing NCBI-disease")
process_ner_dataset('data/NCBI-disease/train.tsv', 'data/NCBI-disease/train.json')
process_ner_dataset('data/NCBI-disease/dev.tsv', 'data/NCBI-disease/dev.json')
process_ner_dataset('data/NCBI-disease/test.tsv', 'data/NCBI-disease/test.json')

print("Processing JNLPBA")
process_ner_dataset('data/JNLPBA/train.tsv', 'data/JNLPBA/train.json')
process_ner_dataset('data/JNLPBA/dev.tsv', 'data/JNLPBA/dev.json')
process_ner_dataset('data/JNLPBA/test.tsv', 'data/JNLPBA/test.json')

print("Processing RE Data")
print("Processing ChemProt")
process_chemprot('data/chemprot/train.tsv', 'data/chemprot/train.json')
process_chemprot('data/chemprot/dev.tsv', 'data/chemprot/dev.json')
process_chemprot('data/chemprot/test.tsv', 'data/chemprot/test.json')
print("Processing DDI" )
process_ddi('data/DDI/train.tsv', 'data/DDI/train.json')
process_ddi('data/DDI/dev.tsv', 'data/DDI/dev.json')
process_ddi('data/DDI/test.tsv', 'data/DDI/test.json')
print("Processing GAD")
process_GAD('data/GAD/train.tsv', 'data/GAD/train.json')
process_GAD('data/GAD/dev.tsv', 'data/GAD/dev.json')
process_GAD('data/GAD/test.tsv', 'data/GAD/test.json')


print("Processin Question Answering)")
print("processing pubmedqa")
process_pubmedqa('data/pubmedqa/pqal_fold0/train_set.json', 'data/pubmedqa/train.json')
process_pubmedqa('data/pubmedqa/pqal_fold0/dev_set.json', 'data/pubmedqa/dev.json')
process_pubmedqa('data/pubmedqa/test_set.json', 'data/pubmedqa/test.json')


print("Processing BioASQ")
process_bioasq('data/bioasq/train.tsv', 'data/bioasq/train.json')
process_bioasq('data/bioasq/dev.tsv', 'data/bioasq/dev.json')
process_bioasq('data/bioasq/test.tsv', 'data/bioasq/test.json')

```
